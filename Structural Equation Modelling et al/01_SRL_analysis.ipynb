```python
def CronbachAlpha(itemscores):
    itemscores = np.asarray(itemscores)
    itemvars = itemscores.var(axis=1, ddof=1)
    tscores = itemscores.sum(axis=0)
    nitems = len(itemscores)

    return nitems / (nitems-1.) * (1 - itemvars.sum() / tscores.var(ddof=1))
```

### SRL Questionnare response patterns

```python
np.random.seed(1)
from scipy.stats import shapiro

srl_summary = pd.DataFrame()
cols_for_correlation = [None]*20
for i in range (1,21):
    col = "I"+str(i)
    
    this_item = srl_data[col].tolist()
    stat, p = shapiro(this_item)
    Res = '< %.3f' % (p)
        
    srl_details = pd.DataFrame({'Item': col
                                ,'Mean':round(srl_data[col].mean(),2)
                                ,'SD':round(srl_data[col].std(),2)
                                ,'Skew':round(srl_data[col].skew(),2)
                                ,'Kurtosis':round(srl_data[col].kurt(),2)
                                ,'Normality (P-value)':Res
                              }, index=[i-1])
    srl_summary = pd.concat([srl_summary,srl_details])
    cols_for_correlation[i-1] = col
srl_summary
```

```python
from scipy.stats import pearsonr

def calculate_pvalues(df):
    df = df.dropna()._get_numeric_data()
    dfcols = pd.DataFrame(columns=df.columns)
    pvalues = dfcols.transpose().join(dfcols, how='outer')
    for r in df.columns:
        for c in df.columns:
            pvalues[r][c] = round(pearsonr(df[r], df[c])[1], 4)
    return pvalues

srl_corr = srl_data[cols_for_correlation]
corr = srl_corr.corr()

pval = calculate_pvalues(srl_corr) 
r1 = corr.applymap(lambda x: 0)
rho = corr.mask(pval>0.05,r1)
rho

show_annot_array = pval < 0.05
annot_array = np.zeros_like(show_annot_array, dtype=np.bool)
annot_array[np.triu_indices_from(show_annot_array)] = True

mask = np.zeros_like(rho, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
fig, ax = plt.subplots(figsize=(15, 12))
# Draw the heatmap with the mask and correct aspect ratio
cmap = sns.cubehelix_palette(50, hue=0.00, rot=0, light=1.0, dark=0,as_cmap=True)

# quick fix
sns.heatmap(show_annot_array, mask=mask, cmap=cmap, vmax=1.,vmin=0.0,
                       annot=True,
            square=True, linewidths=.1, cbar_kws={"shrink": 0.8})

show_annot_array = ax.texts
#plt.show()
plt.close()
```

```python
plt.style.use('seaborn-whitegrid')
sns.set_context("notebook", font_scale=1.2)
sns.set_style("whitegrid",{"font.family": ["Corbel"]})
fig, ax = plt.subplots(figsize=(15, 12))


corr_plot = sns.heatmap(rho, mask=mask, cmap=cmap, vmax=1.,vmin=0.,
                       annot=True,
            square=True, linewidths=.1, cbar_kws={"shrink": 0.9})

ax.set_title("SRL Items Correlation").set_fontsize('18')

for text, show_annot in zip(ax.texts,(bool(int(row.get_text())) for row in show_annot_array)):
    text.set_visible(show_annot)
    
ax.set_ylim(sorted(ax.get_xlim(), reverse=True))
plt.show()
```

The Kruskal-Wallis test is a nonparametric version of the one-way analysis of variance test (ANOVA). This test can be used to determine whether more than two independent samples have a different distribution. It can be thought of as the generalization of the Mann-Whitney U test.

The Mann-Whitney U test is a nonparametric statistical significance test for determining whether two independent samples were drawn from a population with the same distribution.

The Wilcoxon signed ranks test is a nonparametric statistical procedure for comparing two samples that are paired, or related. The parametric equivalent to the Wilcoxon signed ranks test goes by names such as the Studentâ€™s t-test, t-test for matched pairs, t-test for paired samples, or t-test for dependent samples.

The Levene test tests the null hypothesis that all input samples are from populations with equal variances

```python
np.random.seed(1)
from scipy.stats import kruskal,levene
from pingouin import kruskal as pg_kruskal

for i in SRL_manova_transpose['SRL Skill'].unique().tolist():
    this_skill = SRL_manova_transpose[SRL_manova_transpose['SRL Skill']==i]
    
    low_srl = this_skill.Mean[this_skill.SRL =='Low SRL profile'].tolist()
    average_srl = this_skill.Mean[this_skill.SRL =='Average SRL profile'].tolist()
    above_average_srl = this_skill.Mean[this_skill.SRL =='Above Average SRL profile'].tolist()
    high_srl = this_skill.Mean[this_skill.SRL =='High SRL profile'].tolist()
        
    stat, p = kruskal(low_srl, average_srl, above_average_srl, high_srl)
    print(i + ', Kruskal, Statistics=%.3f, p=%.3f' % (stat, p))   
    stat, p = levene(low_srl, average_srl, above_average_srl, high_srl)
    print(i + ", Levene's Test, Statistics=%.3f, p=%.3f" % (stat, p))
    print()    
```
