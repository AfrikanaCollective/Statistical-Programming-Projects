### Get study demographic info for participants 

```python
def flatten_fill(user_data):
    
    Session = user_data.Session.dropna().min()
    
    email = user_data.Email.dropna().unique()
    email = ''.join(email)
    
    check_mobile = user_data.MobileNumber.dropna().unique()
    mobile = np.nan
    
    if('Empty' not in check_mobile):
        check_mobile = ''.join(check_mobile)
        mobile = check_mobile[0:2]
        if(len(mobile)==0):
            mobile = np.nan
    
    check_clincian = user_data.Doctor.dropna().unique()
    clinician = False
    
    if('Yes' in check_clincian):
        clinician = True
        
    check_etat = user_data.ETAT.dropna().unique()
    etat = False
    
    if('Yes' in check_etat):
        etat = True
    
    
    study_group = user_data.StudyGroup.dropna()
    study_group = study_group.astype(str).str.strip()
    study_group = study_group.unique()
    study_group = ''.join(study_group)    
    
    check_consent = user_data.ConsentGiven.dropna().unique()
    consent = True
    
    if('true' not in check_consent):
        consent = False    
    
    
    if not consent:
        user_details = pd.DataFrame({'Session':Session
                                     ,'Email':email
                                     ,'Country':mobile
                                     ,'Clinician':clinician
                                     ,'ETAT':etat
                                     ,'StudyGroup':study_group
                                     ,'ConsentGiven':consent
                                     ,'Institution':np.nan
                                     ,'Age':np.nan
                                     ,'Cadre':np.nan
                                     ,'Level':np.nan
                                     ,'Experience':np.nan
        }, index=user_data.User.unique())
            
    else:
        institution = user_data.Institution.dropna().unique()
        age = user_data.Age.dropna().unique()
        cadre = user_data.Cadre.dropna().unique()
        level = user_data.Level.dropna().unique()
        experience = user_data.Experience.dropna().unique()
        
        if(len(institution) == 0):
            institution = np.nan
        else:
            institution = ''.join(institution)
        if(len(age) == 0):
            age = np.nan
        else:
            age = ''.join(age)
        if(len(cadre) == 0):
            cadre = np.nan
        else:
            cadre = ''.join(cadre)
        if(len(level) == 0):
            level = np.nan
        else:
            level = ''.join(level)
        if(len(experience) == 0):
            experience = np.nan
        else:
            experience = ''.join(experience)          
            
        
        user_details = pd.DataFrame({'Session':Session
                                     ,'Email':email
                                     ,'Country':mobile
                                     ,'Clinician':clinician
                                     ,'ETAT':etat
                                     ,'StudyGroup':study_group
                                     ,'ConsentGiven':consent
                                     ,'Institution':institution
                                     ,'Age':age
                                     ,'Cadre':cadre
                                     ,'Level':level
                                     ,'Experience':experience
        }, index=user_data.User.unique())
    return user_details    
    

users_list=all_users.groupby('User').apply(flatten_fill)
users_list.reset_index(inplace=True)
users_list = users_list.drop('level_1',axis=1)


#Users missing record on users node
aberrant = pd.DataFrame({'Session': datetime.strptime('2019-03-06 17:37:05','%Y-%m-%d %H:%M:%S')
             ,'User':'703D7C0244D777F3EF9D756918231113'
             ,'Email':'Empty'
             ,'Country':'ke'
             ,'Clinician':True
             ,'ETAT':False
             ,'StudyGroup':1
             ,'ConsentGiven':True
             ,'Institution':np.nan
             ,'Age':np.nan
             ,'Cadre':np.nan
             ,'Level':np.nan
             ,'Experience':np.nan
        }, index=[users_list.shape[0]])
users_list = pd.concat([users_list,aberrant])

users_list = users_list.sort_values(by=['Session'],ascending=[True])
users_list = users_list.reset_index(drop=True)
users_list.loc[(users_list.Clinician==False)&(users_list.Cadre.isin(['Doctor','Nurse','Clinical Officer','DoctorNurse'])),['Clinician']]=True

users_list.tail()
```

```python
neonate_play = db.child("resus").child("play").get()
play_data = pd.DataFrame()
for play_node in neonate_play.each():
    user_id = play_node.key()
    if(user_id in users_exclude):
        continue
    else:        
        for key,entries in play_node.val().items():
            if(key=='scenario_a'):                
                for session_key,play_session in entries.items():
                    session = None
                    try:
                        session = datetime.strptime(session_key, '%b %d, %Y, %I:%M:%S %p')
                    except:
                        continue
                    for _,play_stats in play_session.items():
                        stats = play_stats.split(',')
                        stats[0]=stats[0].replace("Question","")
                        if(len(stats)>4):
                            row_cols = ["User","Session","Question","Try","Time","Correct","Feedback"]
                            row_vals=[user_id,session]
                            row_vals.extend(stats)                
                            play_row = pd.DataFrame.from_records([row_vals],columns=row_cols)
                            play_data = pd.concat([play_data,play_row])
            
play_data.Question=play_data.Question.astype(int)
play_data = play_data[play_data.Question != 6]
play_data.Try=play_data.Try.astype(int)
play_data.Correct=play_data.Correct.astype(int)
play_data.Time=play_data.Time.astype(float)
play_data.Time=play_data.Time.round(3)

#play_data = play_data[play_data.User.isin(users_list.User)]
play_data.to_csv(ROOT_DATA+'play_data_raw.csv')
```

```python
play_data['Repeated'] = 0.0
play_data['I'] = 1.0
play_data.Repeated = play_data.groupby(['User','Session','Question','Try']).I.cumcount()
play_data_repeated = play_data.copy()
play_data.drop('I',axis=1,inplace=True)
play_data_repeated.drop('I',axis=1,inplace=True)
play_data_repeated.reset_index(inplace=True,drop=True)
play_data.reset_index(inplace=True,drop=True)
play_data_repeated.tail()
```

```python
new_session_start = play_data_repeated.groupby(['User','Session','Repeated']).Time.sum()
new_session_start = new_session_start.to_frame()
new_session_start.reset_index(inplace=True)
new_session_start['Next_Session'] = new_session_start.Session
new_session_start.tail()
```

```python
for index,row in new_session_start.iterrows():
    current_repeat, user, current_session = row['Repeated'], row['User'], row['Session']
    previous_repeat = current_repeat-1
    if(current_repeat == 1):
        last_session = new_session_start.Session[(new_session_start.User==user)&(new_session_start.Repeated==previous_repeat)&(new_session_start.Session==current_session)].tolist()
        last_session = list(set(last_session))[0]
        last_time = new_session_start.Time[(new_session_start.User==user)&(new_session_start.Repeated==previous_repeat)&(new_session_start.Session==current_session)].tolist()
        last_time = list(set(last_time))[0]
        new_session = last_session + timedelta(seconds=np.ceil(last_time)) + timedelta(seconds=3)
        new_session_start.loc[index,['Next_Session']] = new_session
    
    elif(current_repeat > 1):
        last_session = new_session_start.Next_Session[(new_session_start.User==user)&(new_session_start.Repeated==previous_repeat)&(new_session_start.Session==current_session)].tolist()
        last_session = list(set(last_session))[0]
        last_time = new_session_start.Time[(new_session_start.User==user)&(new_session_start.Repeated==previous_repeat)&(new_session_start.Session==current_session)].tolist()
        last_time = list(set(last_time))[0]
        new_session = last_session + timedelta(seconds=np.ceil(last_time)) + timedelta(seconds=3)
        new_session_start.loc[index,['Next_Session']] = new_session
    
    
new_session_start.drop('Time',axis=1,inplace=True)
play_data_fixed = pd.merge(play_data,new_session_start,how='left',on=['User','Session','Repeated'])
play_data_fixed.Session = play_data_fixed.Next_Session
play_data_fixed.drop(['Repeated','Next_Session'],axis=1,inplace=True)
play_data_test = play_data_fixed.groupby(['User','Session','Question','Try']).first()
play_data_test.reset_index(inplace=True)

play_data = play_data_fixed

play_data.reset_index(drop=True,inplace=True)
play_data=play_data.sort_values(by=['Session','User','Question'],ascending=[True,False,True])
play_data.head()
```

>When testing the data, the cells should be counts of cases and not percentages. It is okay to convert to percentages after testing the data

2. Each participant contributes to only one cell within the Chi-square table
3. The groups being tested must be independent

Don't use the value above, it contravenes these rules

```python
gap_data = play_data[['User','Session','Time']]
session_trackers = play_data[['User','Session','Session_Complete','Complete_Plays']]
session_trackers = session_trackers.drop_duplicates(['User','Session','Session_Complete','Complete_Plays'])
group_data = users_list[['User','StudyGroup']]
gap_data = (gap_data.groupby(['User','Session']).Time.sum()).to_frame()
gap_data.reset_index(inplace=True)
gap_data['Time_Difference'] = gap_data.groupby('User').Session.transform(lambda x: ((x -x.shift()).dt.total_seconds()))
gap_data['Gap']=np.nan

def cumilate_gap(user_data):
    user_data['Gap'] = user_data.Time_Difference + user_data.Time.shift()
    return(user_data)
    
gap_data = gap_data.groupby('User').apply(cumilate_gap)
gap_data.drop(['Time','Time_Difference'],axis=1,inplace=True)
gap_data.Gap = gap_data.Gap/60.0

gap_data['Gap_Type'] = 'None'
gap_data.loc[gap_data.Gap <= 60.0,['Gap_Type']] = '<= 1 Hour'
gap_data.loc[(gap_data.Gap > 60.0) & (gap_data.Gap <= 1440.0),['Gap_Type']] = '<= 1 Day'
gap_data.loc[(gap_data.Gap > 1440.0) & (gap_data.Gap <= 10080.0),['Gap_Type']] = '<= 1 Week'
gap_data.loc[(gap_data.Gap > 10080.0) & (gap_data.Gap <=  40320.0),['Gap_Type']] = '<= 1 Month'
gap_data.loc[(gap_data.Gap > 40320.0),['Gap_Type']] = '> 1 Month'


gap_data.loc[gap_data.Gap.isnull(),['Gap']]=0.01
gap_data.Gap = np.log(gap_data.Gap)

secondary_analysis_play_data = play_data[['User','Session','Question','Try','Time','Feedback','Correct']]
secondary_analysis_play_data['I'] = 1
secondary_analysis_play_data['Opportunity'] = secondary_analysis_play_data.groupby(['User','Question']).I.cumcount()+1
secondary_analysis_play_data.drop(['I'],axis=1,inplace=True)
secondary_analysis_play_data = pd.merge(secondary_analysis_play_data,gap_data,how='left',on=['User','Session'])
secondary_analysis_play_data = pd.merge(secondary_analysis_play_data,group_data,how='left',on=['User'])
secondary_analysis_play_data.StudyGroup = secondary_analysis_play_data.StudyGroup.astype(str)
secondary_analysis_play_data.loc[secondary_analysis_play_data.StudyGroup=='2',['StudyGroup']] = 'Experiment'
secondary_analysis_play_data.loc[secondary_analysis_play_data.StudyGroup=='1',['StudyGroup']] = 'Control'
secondary_analysis_play_data.rename(columns={'StudyGroup':'Group'},inplace=True)

session_ids  =  play_data[['User','Session']]
session_ids = session_ids.drop_duplicates(['User','Session'])
session_ids['I'] = 1.0
session_ids['Cycle'] = session_ids.groupby(['User']).I.cumcount()+1
session_ids.drop('I',axis=1,inplace=True)
secondary_analysis_play_data = pd.merge(secondary_analysis_play_data,session_ids,how='left',on=['User','Session'])
secondary_analysis_play_data = pd.merge(secondary_analysis_play_data,session_trackers,how='left',on=['User','Session'])

srl_data_mutinom = pd.read_csv('D:\DPhil - University of Oxford\Reports\SRL\multinom.csv')
srl_data_mutinom = srl_data_mutinom[['User','SRL']]
srl_data_mutinom = srl_data_mutinom[~(srl_data_mutinom.User.isnull())]

srl_data_mutinom.SRL = srl_data_mutinom.SRL.astype(str)

srl_data_mutinom.loc[srl_data_mutinom.SRL=='1',['SRL']] = 'Low SRL profile'
srl_data_mutinom.loc[srl_data_mutinom.SRL=='2',['SRL']] = 'Average SRL profile'
srl_data_mutinom.loc[srl_data_mutinom.SRL=='3',['SRL']] = 'Above Average SRL profile'
srl_data_mutinom.loc[srl_data_mutinom.SRL=='4',['SRL']] = 'High SRL profile'

secondary_analysis_play_data = pd.merge(secondary_analysis_play_data,
                                        srl_data_mutinom,how='left',on=['User'])

secondary_analysis_play_data.head()
```

```python
from scipy import stats

users_list_use = user_background
users_list_use = users_list_use[~(users_list_use.Cadre.isnull())] 

users_summary=users_list_use[['Clinician','Cadre','Level','Age','Experience','ETAT','StudyGroup','No_Session','No_Session_Complete']]
users_summary=users_summary[users_summary.Clinician]
users_summary.loc[users_summary.Level=='General OfficerStudentIntern',['Level']]='Student'
users_summary.loc[users_summary.Level=='StudentIntern',['Level']]='Student'
users_summary.loc[users_summary.Cadre=='Clinical OfficerDoctor',['Cadre']]='Clinical Officer'
users_summary.loc[users_summary.Age.isnull(),['Age']]=0
users_summary.Age = users_summary.Age.astype(str)
users_summary.Age = users_summary.Age.str[:2]
users_summary.Age = users_summary.Age.astype(int)

users_summary.loc[users_summary.Experience.isnull(),['Experience']]=0
users_summary.Experience = users_summary.Experience.astype(str)
users_summary.Experience = users_summary.Experience.str[:2]
users_summary.Experience = users_summary.Experience.astype(int)

def remove_na(column_data,col,frame_cont):
    column_data = column_data[~(column_data[col].isnull())]
    column_data[col] = column_data[col].astype(float) 
    column_data = column_data[column_data[col] > 0]
    
    #Mean
    Ctrl_List = column_data[col][column_data['StudyGroup']==1]
    Exp_List = column_data[col][column_data['StudyGroup']==2]
    p_value = np.round(stats.ttest_ind(Ctrl_List.tolist(),Exp_List.tolist(), equal_var = False),3)[1]
        
    res = column_data.groupby('StudyGroup')[col].mean()
    res = res.to_frame()
    res.reset_index(inplace=True)
    res=res[res.StudyGroup != -1]
    res[col]=round(res[col],2)
    res['Indicator'] = 'Mean'
    
    #SD
    res_sd = column_data.groupby('StudyGroup')[col].std()
    res_sd = res_sd.to_frame()
    res_sd.reset_index(inplace=True)
    res_sd=res_sd[res_sd.StudyGroup != -1]
    res_sd[col]=round(res_sd[col],2)
    res_sd['Indicator'] = 'SD'
    
    #Count
    res_count = column_data.groupby('StudyGroup')[col].count()
    res_count = res_count.to_frame()
    res_count.reset_index(inplace=True)
    res_count = res_count[res_count.StudyGroup != -1]
    res_count[col]=res_count[col].astype(int)
    res_count['Indicator'] = 'N'
        
    res = pd.concat([res,res_sd,res_count])
    
    res = res.pivot_table(col,['Indicator'],'StudyGroup')
    res.rename(columns={1:"Ctrl",2:"Exp"},inplace=True)
    
    #print(stats.ttest_ind(rvs1,rvs2, equal_var = False))
    
    results_col = pd.DataFrame({
        'Column':col,
        'Control': str(res['Ctrl'][0])+" ("+str(res['Ctrl'][2])+", "+str(int(res['Ctrl'][1]))+")",
        'Experiment': str(res['Exp'][0])+" ("+str(res['Exp'][2])+", "+str(int(res['Exp'][1]))+")",
        'P-Value': p_value
    },index=[0])
    frame_cont = pd.concat([frame_cont,results_col])
    
    return frame_cont

def remove_na_new(column_data,col,frame_cont):
    column_data = column_data[~(column_data[col].isnull())]
    column_data[col] = column_data[col].astype(float) 
    column_data = column_data[column_data[col] > 0]
    
    #Mean
    Ctrl_List = column_data[col][column_data['Group']=='Control']
    Exp_List = column_data[col][column_data['Group']=='Experiment']
    p_value = np.round(stats.ttest_ind(Ctrl_List.tolist(),Exp_List.tolist(), equal_var = False),3)[1]
        
    res = column_data.groupby('Group')[col].mean()
    res = res.to_frame()
    res.reset_index(inplace=True)
    res[col]=round(res[col],2)
    res['Indicator'] = 'Mean'
    
    #SD
    res_sd = column_data.groupby('Group')[col].std()
    res_sd = res_sd.to_frame()
    res_sd.reset_index(inplace=True)
    res_sd[col]=round(res_sd[col],2)
    res_sd['Indicator'] = 'SD'
    
    res = pd.concat([res,res_sd])
    
    res = res.pivot_table(col,['Indicator'],'Group')
    res.rename(columns={1:"Control",2:"Experiment"},inplace=True)
    
    
    #print(stats.ttest_ind(rvs1,rvs2, equal_var = False))
    
    results_col = pd.DataFrame({
        'Column':col,
        'Control': str(res['Control'][0])+" ("+str(res['Control'][1])+")",
        'Experiment': str(res['Experiment'][0])+" ("+str(res['Experiment'][1])+")",
        'P-Value': p_value
    },index=[0])
    frame_cont = pd.concat([frame_cont,results_col])
    
    return frame_cont

def get_summary_frame(users_summary):
    
    frame_cont = pd.DataFrame()
    cols_summary_cont = ['Age','Experience','No_Session','No_Session_Complete']
    for i in cols_summary_cont:
        frame_cont = remove_na(users_summary[['StudyGroup',i]],i,frame_cont)  
        
    frame_cont.reset_index(inplace=True,drop=True)
    
    users_summary['LevelI'] = 0
    users_summary.loc[~(users_summary.Level.isnull()),['LevelI']] = 1
    
    level_frame = users_summary.groupby(['StudyGroup','Cadre'])['LevelI'].sum()
    level_frame = level_frame.to_frame()
    level_frame.reset_index(inplace=True)
    res = level_frame.pivot_table('LevelI',['Cadre'],'StudyGroup')
    res.reset_index(inplace=True)
    res.rename(columns={'Cadre':'Column',1:"Control",2:"Experiment"},inplace=True)
    
    users_summary['CadreI'] = 0
    users_summary.loc[~(users_summary.Level.isnull()),['CadreI']] = 1
    
    lvl_frame = users_summary.groupby(['StudyGroup','Level'])['CadreI'].sum()
    lvl_frame = lvl_frame.to_frame()
    lvl_frame.reset_index(inplace=True)
    res_lvl= lvl_frame.pivot_table('CadreI',['Level'],'StudyGroup')
    res_lvl.reset_index(inplace=True)
    res_lvl.rename(columns={'Level':'Column',1:"Control",2:"Experiment"},inplace=True)
    res_lvl
    
    etat_frame = (users_summary.groupby(['StudyGroup'])['ETAT'].count()).to_frame()
    etat_frame.reset_index(inplace=True)
    etat_frame = etat_frame[etat_frame.StudyGroup!=-1]
    etat_frame['Indicator'] = 'N'
    etat_frame = etat_frame.pivot_table('ETAT',['Indicator'],'StudyGroup')
    etat_frame.reset_index(inplace=True)
    etat_frame.rename(columns={'Indicator':'Column',1:"Control",2:"Experiment"},inplace=True)
    
    etat_done_frame = (users_summary.groupby(['StudyGroup'])['ETAT'].sum()).to_frame()
    etat_done_frame.reset_index(inplace=True)
    etat_done_frame = etat_done_frame[etat_done_frame.StudyGroup!=-1]
    etat_done_frame['Indicator'] = 'Done_ETAT'
    etat_done_frame = etat_done_frame.pivot_table('ETAT',['Indicator'],'StudyGroup')
    etat_done_frame.reset_index(inplace=True)
    etat_done_frame.rename(columns={'Indicator':'Column',1:"Control",2:"Experiment"},inplace=True)
    etat_done_frame

    frame_cont = pd.concat([frame_cont,res,res_lvl,etat_done_frame,etat_frame])
    frame_cont.reset_index(inplace=True,drop=True)
    frame_cont.rename(columns={'Column':'Indicator'},inplace=True)
    frame_cont.reset_index()
    
    return frame_cont

all_summary = get_summary_frame(users_summary)
complete_users_summary = users_summary[users_summary.No_Session_Complete > 1.0]
complete_summary = get_summary_frame(complete_users_summary)
incomplete_users_summary = users_summary[users_summary.No_Session_Complete < 2.0]
incomplete_summary = get_summary_frame(incomplete_users_summary)
```
